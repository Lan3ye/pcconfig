```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Path = "./data-staging/video-card.json"
Data = pd.read_json(Path)

# Removing all rows where price is null
Data.dropna(subset=['price'], inplace=True)

plt.scatter(Data['memory'], Data['price'])
plt.show()
```

```python
X = Data['memory'].values.reshape(-1, 1)
y = Data['price']

model = LinearRegression()

model.fit(X,y)

prediction = model.predict(X)

plt.scatter(X, y, label='original data')
plt.plot(X, prediction, color='red', label=f"""m = {model.coef_[0].round(2)} \n n = {model.intercept_.round(2)}""")
plt.title('Linear Regression: Video Memory vs Price')
plt.xlabel('Video Memory')
plt.ylabel('Price')
plt.legend()
plt.savefig('gpu.png', dpi=300)
plt.show()
```
